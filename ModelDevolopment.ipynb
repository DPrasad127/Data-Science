```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Ensure the dataset is downloaded if it's not already present
file_path= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
!wget -O usedcars.csv {file_path}

# Re-load df for self-contained cell execution
df = pd.read_csv("usedcars.csv")

# Define X and y for this cell
X = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
y = df['price']

# Define 'pr' within this cell to ensure it's available
pr = PolynomialFeatures(degree=2)

# Transform X to polynomial features
Z_pr = pr.fit_transform(X)

# Ensure Z_pr is of float type
Z_pr = Z_pr.astype(float)

# Initialize the LinearRegression model for 'pipe'
pipe = LinearRegression()

# Fit the linear model to the polynomial features (Z_pr) and the target variable (y)
pipe.fit(Z_pr, y)
```

```python
ypipe=pipe.predict(Z_pr)
ypipe[0:4]
```

```python
Input=[('scale',StandardScaler()),('model',LinearRegression())]

pipe=Pipeline(Input)

pipe.fit(Z_pr,y)

ypipe=pipe.predict(Z_pr)
ypipe[0:10]
```

```python
## calcualte the r squarwed error
```

```python
print('The R-square is: ', pipe.score(Z_pr, y))
```

```python
import pandas as pd

# Get the mean of the other features from the training data (df)
mean_horsepower = df['horsepower'].mean()
mean_curb_weight = df['curb-weight'].mean()
mean_engine_size = df['engine-size'].mean()

# Create a DataFrame for prediction with 4 features
# Assuming new_input is intended for 'highway-mpg'
num_predictions = new_input.shape[0]
prediction_data = pd.DataFrame({
    'horsepower': [mean_horsepower] * num_predictions,
    'curb-weight': [mean_curb_weight] * num_predictions,
    'engine-size': [mean_engine_size] * num_predictions,
    'highway-mpg': new_input.flatten() # Use the values from new_input for highway-mpg
})

yhat = lm.predict(prediction_data)
yhat[0:5]
```
